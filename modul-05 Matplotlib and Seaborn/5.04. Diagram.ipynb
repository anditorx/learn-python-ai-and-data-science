### Membuat scarlet plot

```python
x = np.array([1,3,2,4,8,15,9,10,11,10])
y = np.array([2,4,1,12,4,13,3,22,12,13])

plt.scatter(x, y)
plt.show()
```

```python
x = np.array([1,3,2,4,8,15,9,10,11,10])
y = np.array([2,4,1,12,4,13,3,22,12,13])
size = np.array([31,12,13,42,15,82,92,10,19,28])

plt.scatter(x, y, s=size, color="#27ae60")
plt.show()
```

### Membuat bar chart

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,4,10])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,4,10])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,4,10])

plt.bar(x,y, color="#27ae60")
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,4,10])

plt.bar(x,y, width=0.1, color="#27ae60")
plt.show()
```

### Membuat pie chart

```python
x = np.array([23,12,13,39,34])
plt.pie(x)
plt.show()
```

```python
colors = ["yellow", "hotpink", "b", "#81ecec"]
x = np.array([23,12,13,39,34])
plt.pie(x, colors=colors)
plt.show()
```

```python
labels=["Bananas", "Apple", "Watermelon", "Tomat"]
colors = ["yellow", "hotpink", "b", "#81ecec"]

x = np.array([23,12,13,39])

plt.pie(x, labels=labels, colors=colors)
plt.show()
```

```python
labels=["Bananas", "Apple", "Watermelon", "Tomat"]
colors = ["yellow", "hotpink", "b", "#81ecec"]
explode = [0, 0, 0, 0.1]

x = np.array([23,12,13,39])

plt.pie(x, labels=labels, colors=colors, explode=explode)
plt.show()
```

```python
labels=["Bananas", "Apple", "Watermelon", "Tomat"]
colors = ["yellow", "hotpink", "b", "#81ecec"]
explode = [0, 0, 0, 0.1]

x = np.array([23,12,6,24])

plt.pie(x, labels=labels, colors=colors, explode=explode)
plt.legend(loc="upper left")
plt.show()
```
